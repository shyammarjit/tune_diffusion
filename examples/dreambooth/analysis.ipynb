```python
import pandas as pd

file_path = '/home/btech/ayush.singh/HSN/tune_diffusion/examples/dreambooth/data.csv'
df = pd.read_csv(file_path)


# Step 3: Filter the DataFrame to select all rows with the desired value
filtered_data = df[df['Dataset_Name'] == 'clock']
filtered_data = filtered_data[filtered_data['diffusion_ver'] == 'sdxl']
sorted_df = filtered_data.sort_values(by='clipi', ascending=False)

# Step 4: Display the filtered data
print(sorted_df)
```

```python
sorted_df = filtered_data.sort_values(by='clipt', ascending=False)
sorted_df
```

```python
columns_to_normalize = ['clipi', 'clipt']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
sorted_df[columns_to_normalize] = scaler.fit_transform(sorted_df[columns_to_normalize])

# Step 4: Calculate a new column 'Normalised' by summing the normalized values of the selected columns
sorted_df['Normalised'] = sorted_df[columns_to_normalize].sum(axis=1)

# Step 5: Sort the DataFrame based on the 'Normalised' column in descending order
sorted_df = sorted_df.sort_values(by='Normalised', ascending=False)

# Step 6: Display the sorted DataFrame
print(sorted_df['Normalised'])
```
