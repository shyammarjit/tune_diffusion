```python
for x in range(1):
    for y in range(1,dataset+1):
        flattened_list = m.values.flatten()
        # Step 3: Replace None values with the desired string
        desired_string = "N/A"  # Replace 'N/A' with any string you want to append
        final_list = [str(val) if val is not None else desired_string for val in flattened_list]
        folder_paths = list(final_list)

        from PIL import Image
        import os
        from PIL import Image, ImageDraw, ImageFont
        iteration = 4
        no_of_lr = 16

        def load_images_from_folders(folder_paths):
            images = []
            for folder_path in folder_paths:
                img_path = os.path.join(folder_path, f"image_{y}.jpg")
                print(img_path)
                if os.path.exists(img_path):
                    try:
                        img = Image.open(img_path)
                        images.append(img)
                    except:
                        print(f"Could not load image from {img_path}")
                else:
                    width, height = images[0].size if images else (100, 100)  
                    blank_image = Image.new('RGB', (width, height), (255, 255, 255))
                    images.append(blank_image)
            return images


        all_images = load_images_from_folders(folder_paths)

        if len(all_images) < iteration*no_of_lr:
            print("Warning: Fewer than 12 images found. Filling with blank images.")
            while len(all_images) < iteration*no_of_lr:
                width, height = all_images[0].size if all_images else (100, 100)  
                blank_image = Image.new('RGB', (width, height), (255, 255, 255))
                all_images.append(blank_image)

        width, height = all_images[0].size  
        combined_image = Image.new('RGB', (width * no_of_lr, height * iteration))
        # Increase the text size by 10 times
        font_size = 50
        font = ImageFont.load_default()  # Using the default system font
        draw = ImageDraw.Draw(combined_image)

        for i in range(iteration):
            for j in range(no_of_lr):
                index = i * no_of_lr + j
                combined_image.paste(all_images[index], (j * width, i * height))
                text = f"Learning rate: {learning_rates[j]}"
                text_width, text_height = draw.textsize(text, font=font)
                text_position = (j * width, i * height + height - text_height)
                
                # Draw the learning rate text on the image
                draw.text(text_position, text, font=font, fill=(0, 0, 0))

        folder_path = f"./images/{type_}/{dataset_name}/{x}/"

        # Check if the folder exists, and create it if it doesn't
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        combined_image.show()
        combined_image.save(f"./images/{type_}/{dataset_name}/{x}/prompt{y}.png")
        print("done")
        
        
```
